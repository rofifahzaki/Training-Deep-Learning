##**Import Library**

```python
#import library 
import tensorflow as tf 
from tensorflow import keras
import numpy as np
from tensorflow.keras.metrics import mean_squared_error
import pandas as pd
import math
```

```python
#Menampilkan versi Tensorflow
tf.__version__
```

##**Tiga Tahap Penting:**
1. Menyusun Arsitektur Model
2. Mengkompilasi Model
3. Training Model


```python
# Menyusun model
model = keras.Sequential([            # Membentuk model squential
  tf.keras.layers.Dense(units=1,             
                        input_shape=[2], 
                        activation='relu')   # Menambahkan layer pada model NN
])

# Compile model
model.compile(loss='mean_squared_error',     # Menambahkan fungsi loss
              optimizer='sgd')               # Menambahkan fungsi optimizer

model.summary()                              # Menampilkan summary model
```

Berdasarkan model summary di atas maka kita sedang membuat model NN dengan satu titik layer. Kemudian hasil parameternya adalah 3, itu berarti terdapat 3 parameter, yaitu w1,w2 dan karena inputnya adalah x1 dan x2, serta b yaitu bias. Berikut adalah visualisasi NN modelnya:

1.png

```python
#Mendefinisikan Data Input (X) & Output (Y)
x=np.array([[1,1],[1,0],[0,1],[0,0]], dtype="int")
y=np.array([[1],[0],[0],[0]], dtype="int")
```

```python
#Menampilkan data X
x
```

```python
#Menampilkan data Y
y
```

```python
#Melatih model (fitting/training)
history = model.fit(x, y, epochs=2000, verbose=2)
```

```python
#Prediksi Model
ypred=model.predict(x)
ypred
```

```python
#Membulatkan Hasil Pemodelan
model.get_weights()
```

```python
mse = mean_squared_error(y, ypred)
mse
```
